```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

im_gray = cv2.imread("lena.jpg", 0)
plt.figure(figsize=(10,10))
plt.imshow(im_gray, cmap="gray")
plt.show()
plt.xlim(0,255)
plt.hist(im_gray.ravel(), 256)
plt.show()
#improving the contrast
im_equ = cv2.equalizeHist(im_gray)
plt.imshow(im_equ, cmap="gray")
plt.show()
plt.xlim(0,255)
plt.hist(im_equ.ravel(), 256)
plt.show()
```

```python
'''
task 2

equalize to improve contrast, can see better different kinds of gray values
contrast improvement
'''
import cv2
import matplotlib.pyplot as plt
import numpy as np

im_gray = cv2.imread("locontrast.jpg", 0)

hist = cv2.calcHist([im_gray],[0],None,[256],[0,256])

equ_img = cv2.equalizeHist(im_gray)
histcv2equ = cv2.calcHist([equ_img],[0],None,[256],[0,256])

my_dpi = 200
fig = plt.figure(figsize=(2, 2), dpi=my_dpi)
ax1 = fig.add_subplot(2, 2, (1,2))
ax1.axis('off')
concatenate = np.hstack((im_gray, equ_img))
ax1.imshow(concatenate, cmap="gray")

ax2 = fig.add_subplot(2, 2, 3)
ax2.axis('off')
ax2.plot(hist)

ax3 = fig.add_subplot(2, 2, 4)
ax3.axis('off')
ax3.plot(histcv2equ)

plt.show()
```

```python
'''
task 3
normalize, stretch the image, increase distance between min max of gray values
'''
import cv2

im = cv2.imread('locontrast.jpg')
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

#Plot color histogram excluding black due to high frequency
norm_img = cv2.normalize(im_rgb, None, 0, 255, cv2.NORM_MINMAX)

histcv2equ = cv2.calcHist([norm_img],[0],None,[256],[0,256])
hist = cv2.calcHist([im_rgb],[0],None,[256],[0,256])

my_dpi = 200
fig = plt.figure(figsize=(2, 2), dpi=my_dpi)
ax1 = fig.add_subplot(2, 2, (1,2))
ax1.axis('off')
concatenate = np.hstack((im_rgb, norm_img))
ax1.imshow(concatenate, cmap="gray")

ax2 = fig.add_subplot(2, 2, 3)
ax2.axis('off')
ax2.plot(histcv2equ)

ax3 = fig.add_subplot(2, 2, 4)
ax3.axis('off')
ax3.plot(hist)
```
