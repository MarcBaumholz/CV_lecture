```python
'''
grayscale HSV image
h, s, v = hsv_img[:, :, 0], hsv_img[:, :, 1], hsv_img[:, :, 2]
plt.imshow(hsv_img)
plt.imshow(h)
plt.imshow(s)
plt.imshow(v)
'''
import cv2
import matplotlib.pyplot as plt
import numpy as np

im = cv2.imread('lena.jpg')
imRGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
# first method
imGray = cv2.cvtColor(imRGB, cv2.COLOR_RGB2GRAY)
plt.imshow(imGray, cmap="gray")
plt.show()

#second method
hsv_img = cv2.cvtColor(imRGB, cv2.COLOR_RGB2HSV)
h,s,v = cv2.split(hsv_img)
s[s <= 255] = 0
mergedHSV = cv2.merge((h,s,v))

plt.imshow(v, cmap="gray")
plt.show()

imRGBBack = cv2.cvtColor(mergedHSV, cv2.COLOR_HSV2RGB)
plt.imshow(imRGBBack)
plt.show()

#third method
rows, columns, channel= imRGB.shape
imGray = np.zeros(shape=(rows, columns), dtype="uint8")
for r in range(rows):
    for c in range(columns):
        imGray[r,c] = (0.298936*imRGB[r,c,0] + 0.587043 * imRGB[r,c,1] + 0.114021 * imRGB[r,c,2])

plt.imshow(imGray, cmap="gray")
plt.show()
```
