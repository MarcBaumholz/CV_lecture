```python
'''
task 1
=> gry => resize => imutils => rotate
'''
import cv2
import imutils
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)


dst = cv2.resize(img,(400,400))
angle = 30
rotated = imutils.rotate(dst, angle, scale=1)
rotated_bound = imutils.rotate_bound(dst, angle)

cv2_rot = cv2.rotate(dst, cv2.cv2.ROTATE_90_CLOCKWISE)


plt.imshow(rotated, cmap="gray")
plt.show()

plt.imshow(rotated_bound, cmap="gray")
plt.show()

plt.imshow(cv2_rot, cmap="gray")
plt.show()

```

```python
'''
Task 3
load 2 different images resize to same
'''
import cv2
import imutils
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

height,width  = img1.shape[:2] #(r,c) h=rows, w=column

img2 = cv2.imread('horse_r.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)


dst = cv2.resize(img2,(width,height))

angle = 30
rotated = imutils.rotate(dst, angle, scale=1)
rotated_bound = imutils.rotate_bound(dst, angle)

plt.imshow(rotated, cmap="gray")
plt.show()

plt.imshow(rotated_bound, cmap="gray")
plt.show()

```

```python
'''
taks 4
    resize and stitch img together
'''
import cv2
import imutils
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

img2 = cv2.imread('trees.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

print(img1.shape)
print(img2.shape)

height1, width1  = img1.shape[:2]
height2, width2  = img2.shape[:2]

if width1 > width2:
    resized = cv2.resize(img2,(width1,height1))
    print(img1.shape)
    print(resized.shape)
    stiched = np.hstack((img1, resized))
else:
    resized = cv2.resize(img1,(width2,height2))
    print(img1.shape)
    print(resized.shape)
    stiched = np.hstack((img2, resized))


plt.imshow(stiched, cmap="gray")
plt.show()
```

```python
'''
practice
overlay
'''
import cv2
import imutils
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('horse_r.png')

height,width = img.shape[:2]

dst = cv2.resize(img2,(width,height))

overlay = cv2.addWeighted(img, 0.7, dst, 0.6, 0.0)

concatenate = np.concatenate((dst, img), axis=1)

plt.imshow(overlay)
plt.show()

plt.imshow(concatenate)
plt.show()
```
