```python
'''
task 1
rgb to binary threshold
'''
import cv2
import numpy as np
from matplotlib import pyplot as plt

imBGR = cv2.imread('trees.png')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)

my_dpi = 200
fig = plt.figure(figsize=(10, 10), dpi=my_dpi)

ax1 = fig.add_subplot(2, 2, 1)
ax1.axis('off')
ret, bw_img = cv2.threshold(imRGB, 127, 255, cv2.THRESH_BINARY)
ax1.imshow(bw_img, cmap="gray")
#Obtain and display Cyan

ax2 = fig.add_subplot(2, 2, 2)
ax2.axis('off')
ret, bw_img_25 = cv2.threshold(imRGB, 63, 255, cv2.THRESH_BINARY)
ax2.imshow(bw_img_25, cmap="gray")

ax3 = fig.add_subplot(2, 2, 3)
ax3.axis('off')
ret, bw_img_75 = cv2.threshold(imRGB, 191, 255, cv2.THRESH_BINARY)
ax3.imshow(bw_img_75, cmap="gray")

ax4 = fig.add_subplot(2, 2, 4)
ax4.axis('off')
ax4.imshow(imRGB, cmap="gray")


```

```python

'''
task 2
gray to binary threshold width adaptive thersholding
'''
import cv2
import numpy as np
from matplotlib import pyplot as plt

imGRAY = cv2.imread('creative-proposal.jpg',0)
thresh_50 = 50*255//100

my_dpi = 200
fig = plt.figure(figsize=(2, 2), dpi=my_dpi)

ax1 = fig.add_subplot(2, 2, 1)
ax1.axis('off')
ret, bw_img = cv2.threshold(imGRAY, 127, 255, cv2.THRESH_BINARY)
ax1.imshow(bw_img, cmap="gray")
#Obtain and display
ax2 = fig.add_subplot(2, 2, 2)
ax2.axis('off')
adap_gaus = cv2.adaptiveThreshold(imGRAY, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 12)
ax2.imshow(adap_gaus, cmap="gray")

ax3 = fig.add_subplot(2, 2, 3)
ax3.axis('off')
adap_mean = cv2.adaptiveThreshold(imGRAY,255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 12)
ax3.imshow(adap_mean,cmap="gray")

ax4 = fig.add_subplot(2, 2, 4)
ax4.axis('off')
ax4.imshow(imGRAY,cmap="gray")
```

```python
'''
task 3
erosion => take a way pixel dilation => we expand pixels
'''
import cv2
import numpy as np
from matplotlib import pyplot as plt

imBGR = cv2.imread('j.png')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

ret, bw_img = cv2.threshold(imRGB, 127, 255, cv2.THRESH_BINARY)
plt.imshow(bw_img)
plt.show()

kernel = np.ones((5,5), np.uint8)
image_eros = cv2.erode(imRGB, kernel)
plt.imshow(image_eros)
plt.show()

img_dilation = cv2.dilate(imRGB, kernel, iterations=1)
plt.imshow(img_dilation)
plt.show()
```

```python
'''
task 4
opening closing
'''
import cv2
import numpy as np
from matplotlib import pyplot as plt

imBGR = cv2.imread('opening_original.png')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

imBGR_c = cv2.imread('closing_original.png')
imRGB_c = cv2.cvtColor(imBGR_c, cv2.COLOR_BGR2RGB)


my_dpi = 200
fig = plt.figure(figsize=(2, 2), dpi=my_dpi)

ax1 = fig.add_subplot(2, 2, 1)
ax1.axis('off')
ax1.imshow(imRGB)

ax2 = fig.add_subplot(2, 2, 2)
ax2.axis('off')
kernel = np.ones((5,5), np.uint8)
opening = cv2.morphologyEx(imRGB, cv2.MORPH_OPEN, kernel)
ax2.imshow(opening)

ax3 = fig.add_subplot(2, 2, 3)
ax3.axis('off')
ax3.imshow(imRGB_c)

ax4 = fig.add_subplot(2, 2, 4)
ax4.axis('off')
closing = cv2.morphologyEx(imRGB_c, cv2.MORPH_CLOSE, kernel)
ax4.imshow(closing)
```

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np


src1 = cv2.imread('lena.jpg')
src2 = cv2.imread('horse_r.png')
src1 = cv2.cvtColor(src1, cv2.COLOR_BGR2RGB)
src2 = cv2.cvtColor(src2, cv2.COLOR_BGR2RGB)

src2 = cv2.resize(src2, src1.shape[1::-1])

print(src2.shape)
# (225, 400, 3)

print(src2.dtype)
# uint8

dst = cv2.bitwise_and(src1, src2)

titles=['Original', 'Mask', 'Result']
images = [src1, src2, dst]

for count in range(3):
    plt.subplot(1,3,count+1)
    plt.imshow(images[count])
    plt.title(titles[count])
    plt.axis('off')


dst = cv2.cvtColor(dst, cv2.COLOR_RGB2BGR)
cv2.imwrite('opencv_bitwise_and.jpg', dst)

```
