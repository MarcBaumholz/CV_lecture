```python
'''
Detect dots in image
'''
import cv2
import numpy as np
import matplotlib.pyplot as plt

im_gray1 = cv2.imread("black_pixels.bmp")
im_gray = cv2.cvtColor(im_gray1, cv2.COLOR_BGR2GRAY)
plt.imshow(im_gray, cmap="gray")
plt.show()

#loop over the image with dots
    #write on the new image 3 dimensions
#numpy takes first rows then columns, but a center of a circle takes frst x then y x is columsn y is rowss

height, width = im_gray.shape[:2]
myarray = im_gray1.copy()
rows, columns, ch = myarray.shape
count = 0
for r in range(rows):
    for c in range(columns):
        if im_gray[r,c] == 0:
            count += 1
            cv2.circle(myarray, center=(r,c), radius=10, color=(255,0,0),thickness=2)
label = "No of pixels: "+ str(count)
cv2.putText(myarray, label, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0),2)

plt.imshow(myarray)
plt.show()
```
