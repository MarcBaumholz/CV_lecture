```python
'''
brighten HSV image
'''
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

im = cv2.imread('lena.jpg')
imRGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)  #image read as BGR being displayed incorrectly as RGB
plt.show()
#convert BGR to RGB before display
hsv_image = cv2.cvtColor(imRGB, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_image)
plt.show()

brighten = int(input())
# need to split the image to have access to the channels
h,s,v = cv2.split(hsv_image)
#check if the limit gets exceed
limit = 255 - brighten
print(v.shape)
rows, columns = v.shape
for r in range(rows):
    for c in range(columns):
        if v[r, c] > limit:
            v[r,c] = 255
        else:
            v[r, c] = v[r, c] + brighten
mergedHSV = cv2.merge((h,s,v))
imRGBBack = cv2.cvtColor(mergedHSV, cv2.COLOR_HSV2RGB)
plt.imshow(imRGBBack)
plt.show()

Image.fromarray(hsv_image).save('brightened_image.jpg')


```
