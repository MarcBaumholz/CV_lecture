```python
# scaling, translation, rotation, reflection (flipping)

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.jpg')
rows,cols = img.shape[:2]

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
#Original Image:
plt.title("Original")
plt.show()

#resizing by a factor of 0.5
res = cv2.resize(img,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
plt.imshow(res)
plt.title("Resized by factor of 2")
plt.show()
#OR

#setting pivot point to centre and angle to 90 degrees
M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
dst = cv2.warpAffine(res,M,(cols,rows))
#dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(dst)
plt.title("rotate by 180 degree")
plt.show()

#Translate that head is on bottom right
M = np.float32([[1,0,10],[0,1,10]])
tran = cv2.warpAffine(dst,M,(cols,rows))
#tran = cv2.cvtColor(tran, cv2.COLOR_BGR2RGB)
plt.imshow(tran)
plt.title("head bottom right corner")
plt.show()
```

```python
# of an affine transformation by knowing 3 points and their result
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('buildings.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("head bottom right corner")
plt.show()
rows,cols,ch = img.shape

pts1 = np.float32([[580,90],[760,350],[500,410],[760,700]])
pts2 = np.float32([[0,0],[400,0],[0,300],[400,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img_rgb,M,(400,300))


plt.imshow(img)
plt.title('Input')
plt.show()
plt.imshow(dst)
plt.title('Output')
plt.show()
```
