```python
# of an affine transformation by knowing 3 points and their result
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image

img = cv2.imread('almond_milk.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
rows,cols,ch = img.shape
img_rgb = img_rgb +20
pts1 = np.float32([[840,490],[1130,530],[800,1500],[1070,1350]])
pts2 = np.float32([[0,0],[400,0],[0,300],[400,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img_rgb,M,(400,300))

plt.imshow(dst)
plt.title('Output')
plt.show()
Image.fromarray(dst).save('milk_for_text.jpg')
```
