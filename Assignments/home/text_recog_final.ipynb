```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
import re

img = cv2.imread('text_date.png')

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Adding custom options
'https://muthu.co/all-tesseract-ocr-options/'
custom_config = r'--oem 3 --psm 6'
plt.imshow(img)
plt.show()

text = pytesseract.image_to_string(img, config=custom_config)
text_all = text.split(" ")

r = re.compile("^(0?[1-9]|1[012])[.]\d{4}")
filtered_list = list(filter(r.match, text_all))
print(filtered_list)
```

```python
#shape identified mars

import cv2
from matplotlib import pyplot as plt

im = cv2.imread('2.jpg')
plt.imshow(im)
plt.show()
gray= cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray,(5,5),0)

ret, thresh = cv2.threshold(gray,240,255,0)

img_canny = cv2.Canny(thresh,10,200)

contours,hirarchy = cv2.findContours(img_canny,cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
image = cv2.drawContours(im, contours, -1,(255,0,255),8)

im_rgb= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
plt.show()
```

```python
#shape identified Fairy
import numpy as np
import cv2
from matplotlib import pyplot as plt

im = cv2.imread('Fairy_object.png')
plt.imshow(im)
plt.show()

lower_yellow = np.array([25, 50, 50])
upper_yellow = np.array([32, 255, 255])

hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, lower_yellow, upper_yellow)
kernel = np.ones((8,8),np.uint8)
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap='gray')
plt.show()

contours,hirarchy = cv2.findContours(closing,cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
image = cv2.drawContours(im, contours, -1,(255,0,255),5)
plt.imshow(image)
plt.show()
```

```python
#shape identified mars
import numpy as np
import cv2
from matplotlib import pyplot as plt

im = cv2.imread('mars.jpg')
plt.imshow(im)
plt.show()

low_black=np.array([0,0,0])
high_black=np.array([360,255,100])

hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, low_black, high_black)
kernel = np.ones((4,4),np.uint8)
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap='gray')
plt.show()

contours,hirarchy = cv2.findContours(closing,cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
image = cv2.drawContours(im, contours, -1,(255,0,0),5)
plt.imshow(image)
plt.show()
```

```python
#fairy text recognition
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_rgb = cv2.imread('Fairy_object.png')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('fairy_small.png',0)
w, h = template.shape[::-1]

res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.9
loc = np.where( res >= threshold)
count=0
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
    count+=1
img_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show
print(count)
cv2.imwrite('res.png',img_rgb)



```
